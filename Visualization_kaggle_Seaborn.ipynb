### find all columns in dataset

```
list(df_spotify)
```

### add label in lineplot function itself

```
plt.figure(figsize = (20,10))
sns.lineplot(data= df_spotify["Despacito"],label = "Despacito")
sns.lineplot(data= df_spotify["HUMBLE."],label = "Humble")
plt.title('Humble vs Despacito')
```

## 1.2 Bar Charts

```
!wget https://raw.githubusercontent.com/Himanshu-kashyap/Visualization/master/Datasets/flight_delays.csv -O plane.csv
```

```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```
df_plane = pd.read_csv("plane.csv",index_col = "Month")
```

```
df_plane.head()
```

```
plt.figure(figsize = (12,6))
sns.barplot(x = df_plane.index,y = df_plane["NK"])
```

## 1.3 HeatMap

```
plt.figure(figsize = (10,8))
sns.heatmap(data = df_plane,annot = True)
```

## 1.4 ScatterPlot

```
!wget https://raw.githubusercontent.com/Himanshu-kashyap/Visualization/master/Datasets/insurance.csv -O insurance.csv
```

```
df_insurance = pd.read_csv("insurance.csv")
```

```
df_insurance.head()
```

```
sns.scatterplot(x = df_insurance.age,y = df_insurance.charges,hue= df_insurance.smoker)
```

### adding regression line

```
sns.regplot(x = df_insurance.age,y = df_insurance.charges)
```

### adding multiple regression lines

only have to provide names for columns and data **not** dataframe.col

```
sns.lmplot(x = "age",y = "charges",hue = "smoker",data = df_insurance)
```

```
sns.lmplot(x = "bmi",y="charges",data = df_insurance,hue = "smoker")
```

```
sns.lmplot(x = "bmi",y="charges",data = df_insurance,hue = "children")
```

## 1.5 Swarm 
 used for categorical data vs continuous data 

```
sns.swarmplot(x = df_insurance.smoker, y = df_insurance.charges)
```

```
sns.swarmplot(x = df_insurance["children"], y = df_insurance["charges"])
```

## 1.6 Distribution plots / Histogram

```
sns.distplot(a = df_insurance["children"],kde = False)
```

```
sns.distplot(a = df_insurance["children"],kde = True) # kde = kernel density distribution
```

## 1.7 KDE plot

```
sns.kdeplot(data = df_insurance["children"],shade = True)

# adding 2 or more colms does not work
```

### joint / 2D plot

```
plt.xlabel("Children")
plt.ylabel("Charges")
sns.jointplot(x = df_insurance["children"],y = df_insurance["charges"],kind ="kde")

# need float
```

### color coded histogram

```
plt.figure(figsize = (8,8))
pp = df_insurance

sns.distplot(a = pp.children,label = "Children",kde =False)
sns.distplot(a = pp.bmi,label = "BMI",kde =False)
```
